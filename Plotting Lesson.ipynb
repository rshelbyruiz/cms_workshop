Plotting Lesson

```python
import numpy
```

```python
import os
```

```python
distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
print(distances)
```

```python
headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)
```

```python
print(headers)
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,0],data[:,1])
```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
fig_1 = plt.plot(data[:,0], data[:,1], label=sample)
plt.legend()

plt.savefig('all_samples_300dpi.png', dpi=300)
```

```python

for col in range(1, len(headers)):
    plt.figure()
    fig = plt.plot(data[:,0], data[:,col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstroms)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png', dpi=300)
```
